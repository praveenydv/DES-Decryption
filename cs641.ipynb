```python
def char_to_bin (s):
    mp =  dict()
    mp['f'] = "0000"
    mp['g'] = "0001" 
    mp['h'] = "0010" 
    mp['i']       = "0011" 
    mp['j'] = "0100" 
    mp['k'] = "0101" 
    mp['l'] = "0110" 
    mp['m'] = "0111" 
    mp['n'] = "1000" 
    mp['o'] = "1001" 
    mp['p'] = "1010" 
    mp['q'] = "1011" 
    mp['r'] = "1100" 
    mp['s'] = "1101" 
    mp['t'] = "1110" 
    mp['u'] = "1111" 
    binary = ""
    for i in range (len(s)):
        binary = binary + mp[s[i]]
    return binary; 
    
def bin_to_char (s):
    mp =dict()
    mp["0000"] = "f"; 
    mp["0001"] = "g"; 
    mp["0010"] = "h"; 
    mp["0011"] = "i"; 
    mp["0100"] = "j"; 
    mp["0101"] = "k"; 
    mp["0110"] = "l"; 
    mp["0111"] = "m"; 
    mp["1000"] = "n"; 
    mp["1001"] = "o"; 
    mp["1010"] = "p"; 
    mp["1011"] = "q"; 
    mp["1100"] = "r"; 
    mp["1101"] = "s"; 
    mp["1110"] = "t"; 
    mp["1111"] = "u"; 
    char = "" 
    for i in range (0,16):
        ch=s[i*4:4*(i+1)]
        char=char+mp[ch];

    return char
    
```

```python
def permute(k,ip,n):
    perm = ""; 
    for i in range(0,n):
        a= ip[i] - 1;
        perm = perm + k[a]; 
    
    return perm; 
```

```python
def inpermute(k,ip,n):
    perm = "00000000000000000000000000000000"; 
    s = list(perm)

    for i in range(0,n):
        s[ip[i]-1] = k[i];
    
    return "".join(s);
```
